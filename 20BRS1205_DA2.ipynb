Code segment 2: Plotting Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread(r'/mnt/e/VSC_workspace/signals_theory_da/selfie_image.jpg')
plt.show()
plt.hist(img.ravel(), bins = 256)
plt.title("Histogram of given image")
plt.xlabel("Value (normalized) ")
plt.ylabel("Frequency of pixel occurance")
```

Code segment 3 : Gaussian filtering (smoothing- low pass)

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
from scipy import ndimage # Multi-dimensional image processing 
img=mpimg.imread('/mnt/e/VSC_workspace/signals_theory_da/selfie_image.jpg') 
blurred_face=ndimage.gaussian_filter(img, sigma=1.5) # Multidimensional Gaussian filter 
plt.figure(1) 
plt.imshow(img,cmap='gray') 
plt.figure(2) 
plt.imshow(blurred_face,cmap='gray') 
```

Code segment 4 : Median filtering

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
from scipy import ndimage # Multi-dimensional image processing 
img=mpimg.imread('/mnt/e/VSC_workspace/signals_theory_da/selfie_image.jpg') 
blurred_face=ndimage.median_filter(img,5) # Calculate a multidimensional median filter 
plt.figure(1) 
plt.imshow(img,cmap='gray') 
plt.figure(2) 
plt.imshow(blurred_face,cmap='gray') 
```

Code segment 5: Laplacian filtering 
(sharpening - high pass)

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
from scipy import ndimage # Multi-dimensional image processing 
img=mpimg.imread('/mnt/e/VSC_workspace/signals_theory_da/selfie_image.jpg') 
blurred_face=ndimage.laplace(img) # N-dimensional Laplace filter based on approximate second derivatives 
plt.figure(1) 
plt.imshow(img,cmap='gray') 
plt.figure(2) 
plt.imshow(blurred_face,cmap='gray') 
```
